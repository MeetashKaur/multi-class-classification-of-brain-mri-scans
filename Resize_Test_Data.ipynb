```python
import os
import cv2
f=0
path ="C:/Brain_Multilabel_MRI_dataset/Testing/pituitary_tumor"
for file in os.listdir(path):
    print(file)
    a=cv2.imread(os.path.join(path,file))
    resize_image= cv2.resize(a,(224,224),interpolation = cv2.INTER_NEAREST)
    cv2.imwrite('C:/Brain_Multilabel_MRI_dataset/Testing/resize_pt_test/'+"image"+str(f)+".jpg",resize_image)
    f +=1;
```
